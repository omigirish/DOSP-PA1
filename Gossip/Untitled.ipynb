```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df = pd.read_csv("output.csv")
```

```python
gossip = df.query("Algorithm == 'Gossip'")
plt.figure(figsize=(10,5))
for topology in ['line','2D', 'imp3D', 'full']:
    x = gossip.query(f"Topology=='{topology}'")["Nodes"]
    y = gossip.query(f"Topology=='{topology}'")["ConvergenceTime"]
    plt.plot(x, y, label = topology) 
plt.legend() 
plt.xlabel('Num of Nodes')
plt.ylabel('Convergence Time(s)')
plt.title("Convergence Time vs Num Nodes for Gossip Algorithm")

plt.show()
```

```python
# Filtering the DataFrame for the Gossip algorithm
gossip = df.query("Algorithm == 'Gossip'")

# Setting up subplots for each topology
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15,10))
topologies = ['line', '2D', 'imp3D', 'full']
axes = axes.flatten()

for i, topology in enumerate(topologies):
    x = gossip.query(f"Topology=='{topology}'")["Nodes"]
    y = gossip.query(f"Topology=='{topology}'")["ConvergenceTime"]
    axes[i].plot(x, y, label=topology)
    axes[i].set_title(f"Topology: {topology}")
    axes[i].set_xlabel('Number of Nodes')
    axes[i].set_ylabel('Convergence Time (s)')
    axes[i].legend()
    axes[i].grid()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
